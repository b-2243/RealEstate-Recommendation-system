```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
real_estate_data = pd.read_csv('/content/drive/MyDrive/dataset/uae_properties.csv')
real_estate_data.head()
```

```python
real_estate_data.isnull().sum()
```

```python
real_estate_data = real_estate_data.drop(['addedOn','type','rera'],axis='columns')
```

```python
real_estate_data.iloc[8]
```

```python
real_estate_data.head()
```

```python
original_data = real_estate_data.copy()
```

```python
original_data.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(real_estate_data['displayAddress']).toarray()
tfidf = pd.DataFrame(tfidf_matrix, columns=tfidf_vectorizer.get_feature_names_out())
tfidf.head()
```

```python
label_encoder = LabelEncoder()
real_estate_data['propertyType'] = label_encoder.fit_transform(real_estate_data['propertyType'])
```

```python
scaler = MinMaxScaler()
real_estate_data[['bathrooms', 'bedrooms', 'price']] = scaler.fit_transform(real_estate_data[['bathrooms', 'bedrooms', 'price']])
```

```python
numerical_features = real_estate_data[['bathrooms', 'bedrooms', 'propertyType', 'price']]
numerical_features.head()
```

```python
combined_features = pd.concat([pd.DataFrame(tfidf), pd.DataFrame(numerical_features)], axis=1)
combined_features.shape
```

```python
combined_features.head()
```

```python
cosine_sim = cosine_similarity(combined_features)
```

```python
original_data.iloc[5]
```

```python
def get_recommendations(property_index):
    similarity_scores = list(enumerate(cosine_sim[property_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    top_n = 5
    top_properties = [i[0] for i in similarity_scores[1:top_n + 1]]
    recommendations = real_estate_data.iloc[top_properties]

    recommendations['propertyType'] = label_encoder.inverse_transform(recommendations['propertyType'])
    recommendations[['bathrooms', 'bedrooms', 'price']] = scaler.inverse_transform(recommendations[['bathrooms', 'bedrooms', 'price']])
    return recommendations

recommended_properties = get_recommendations(155)

print(original_data.loc[recommended_properties.index, ['id', 'title', 'displayAddress', 'bathrooms', 'bedrooms', 'propertyType', 'price']])
```

# Recommendation system

```python
def recommend_by_input(displayAddress, bathrooms, bedrooms, propertyType, price):
    input_tfidf = tfidf_vectorizer.transform([displayAddress]).toarray()

    encoded_property_type = label_encoder.transform([propertyType])[0]

    scaled_features = scaler.transform([[bathrooms, bedrooms, price]])[0]

    input_vector = list(input_tfidf[0]) + list(scaled_features[:2]) + [encoded_property_type] + [scaled_features[2]]
    input_vector_df = pd.DataFrame([input_vector])

    similarities = cosine_similarity(input_vector_df, combined_features).flatten()

    top_indices = similarities.argsort()[::-1][1:6]
    recommendations = real_estate_data.iloc[top_indices]

    recommendations['propertyType'] = label_encoder.inverse_transform(recommendations['propertyType'])
    recommendations[['bathrooms', 'bedrooms', 'price']] = scaler.inverse_transform(recommendations[['bathrooms', 'bedrooms', 'price']])

    return original_data.loc[recommendations.index, ['id', 'title', 'displayAddress', 'bathrooms', 'bedrooms', 'propertyType', 'price']]
```

```python
recommended = recommend_by_input(
    displayAddress="Downtown Dubai",
    bathrooms=2,
    bedrooms=3,
    propertyType="apartment",
    price=2500000
)
print(recommended)
```
